```python
# Q-2 write a class calculator which takes a number as input and returns the square, cube and square root of that number
import math


class calculator():
    

    def __init__(self,num):
        self.num=num
    
    def square(self):
        print(f"Square is : {self.num*self.num}")
    
    def cube(self):
        print(f"Cube is : {self.num*self.num*self.num}")

    def square_root(self):
        print(f"Square_root is : {math.sqrt(self.num)}")

x=calculator(5)
x.square()
x.cube()
x.square_root()
```

```python
# Q-4 Add a static method in problem 2 to great the user with hello
 # Q-2 write a class calculator which takes a number as input and returns the square, cube and square root of that number
import math


class calculator():
    

    def __init__(self,num):
        self.num=num
    
    def square(self):
        print(f"Square is : {self.num*self.num}")
    
    def cube(self):
        print(f"Cube is : {self.num*self.num*self.num}")

    def square_root(self):
        print(f"Square_root is : {math.sqrt(self.num)}")
    @staticmethod
    def greet():
        print("Hello User")

x=calculator(5)
x.square()
x.cube()
x.square_root()
x.greet()
```

```python
# Q-5 write a class train which has methods to book a ticket, get status (no of seats) 
# and get fare in information of trains runnings under Indian Railways
import random


class trains:
    def book(self,trainNo,fro,to):
        print(f"Your ticket is booked from {fro} to {to} in train number {trainNo}")
    def get_status(self,trainNo):
        print(f"Train no {trainNo} has {random.randint(0,100)} seats available And running on time")
    def get_fare(self,trainNo,fro,to):
        print(f"Your fare from {fro} to {to} in train number {trainNo} is {random.randint(100,500)  }")

t1=trains()
t1.book(12345,"Delhi","Mumbai")
t1.get_status(12345)
t1.get_fare(12345,"Delhi","Mumbai")
```
