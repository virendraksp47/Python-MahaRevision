```python
# Q-1 creat a classs (2-D vector) and use it to create another class representing 3-D vector.
class Vector2D:
    def __init__(self,i,j):
        self.i=i
        self.j=j
    def display(self):
        print(f"2-D vector: {self.i}i + {self.j}j")
class Vector3D(Vector2D):
    
    def __init__(self,i,j,k):
        super().__init__(i,j)
        self.k=k
    def display(self):
        super().display()
        print(f"3-D vector: {self.i}i + {self.j}j + {self.k}k")

x=Vector3D(5, 3, 2)
x.display()
```
