```python
# Q-1 Write a program to open three files 1.txt, 2.txt, 3.txt if any these files are not
# present , a message without existing the program must be printed prompting the same
try:
    with open('1.txt', 'r') as file1:
        print(file1.read())
except FileNotFoundError:
    print("1.txt does not exist.")
try:
    with open('2.txt', 'r') as file2:
        print(file2.read())
except FileNotFoundError:
    print("2.txt does not exist.")
try:
    with open('3.txt', 'r') as file3:
        print(file3.read())
except FileNotFoundError:
    print("3.txt does not exist.")
```

```python
# Q-2 write a program to print third, fifth and seventh elements from a list using enumerate function
my_list=[10,20,30,40,50,60,70,80]

for index, value in enumerate(my_list):
    if index == 2 or index == 4 or index == 6:
        print(value)
```

```python
# Q-3 write a list comprehension to print a list which contains the multiplication table of a user entered number.
number = int(input("Enter a number: "))
mlt_table=[number*i for i in range(1,11)]
print(mlt_table)
```

```python
# Q-4 write a program to display a/b where a and b are integers. if b=0 display infinite by handling 'ZeroDivisionError'
try:
    n1=int(input("Enter the first number: "))
    n2=int(input("Enter the second number: "))
    result = n1 / n2
    print(f"The result of {n1} divided by {n2} is: {result}")
except ZeroDivisionError:
    print("Infinite")
```

```python
# Q-5 Store the multiplicaton tables generated in program 3 in a file named Tables.txt
n = int(input("Enter a number: "))

mlt_table=[n*i for i in range(1,11)]
with open('tables.txt', 'a') as f:
    f.write(f"Table of {n}: {str(mlt_table)} \n")
```
