```python
# Q-2 write a program to fill in a letter template given below with name and date
letter = '''Dear <|NAME|>,
you are selected!
<|DATE|>'''
name = input("Enter your name: ")
date=input("enter date: ")
print(letter.replace("<|NAME|>",name).replace("<|DATE|>",date))

```
