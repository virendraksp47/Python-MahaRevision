```python
# Q-1 write a program to create a dictionary of hindi words 
#with values as their english translation .
# provide user with a way to look up the translation of a word.
print("Welcome to Hindi to English Dictionary")
hindi_to_english={
    "jaago":"Wake up",
    "paani":"water",
    "dost":"friend",
    "ghar":"home",
}
word=input("Enter hindi word to translate: ")
print("the english translation is : ",hindi_to_english[word])
```
