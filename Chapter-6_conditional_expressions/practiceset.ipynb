```python
# Q-1 write a program to find greatest of four numbers
# entered by the user .
n1=int(input("Enter number 1: "))
n2=int(input("Enter number 2: "))
n3=int(input("Enter number 3: "))
n4=int(input("Enter number 4: "))
if n1>n2 and n1>n3 and n1>n4:
    print("greatest number is n1:",n1)
elif n2>n1 and n2>n3 and n2>n4:
    print("greatest number is n2:",n2)
elif n3>n1 and n3>n2 and n3>n4:
    print("greatest number is n3:",n3)
elif n4>n1 and n4>n2 and n4>n3:
    print("greatest number is n4:",n4)
```

```python
# Q-2 write a program to find out whether a student is pass
# or fail, if it requires total 40% and at least 33% in each subject to pass .
physics=int(input("Enter marks in pysics: "))
chemistry=int(input("Enter marks in chemistry: "))
maths=int(input("Enter marks in maths: "))
if physics>33 and chemistry>33 and maths>33:
    total=physics+chemistry+maths
    percentage=(total/300)*100
    if percentage>40:
        print("Congratulations! You have passed the exam")
    else:
        print("Sorry! You have failed the exam due to low percentage")
else:
    print("Sorry! You have failed the exam due to low marks in one or more subjects")
```
