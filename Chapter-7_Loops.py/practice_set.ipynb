```python
# Q-4
num=int(input("Enter a number: "))
for i in range(2,num):
    if num%i==0:
        print("The number is not prime")
        break
else:    print("The number is prime")
```

```python
# Q-6
num=int(input("Enter a number: "))
product=1
for i in range(1,num+1):
    product*=i
print(f"The factorial of {num} is: {product}")
```

```python
# Q-9
num=int(input("Enter a number: "))
for i in range(1,num+1):
    if i==1 or i==num:
        print("*"*num,end="")
        print("")
    else:
        print("*",end="")
        print(" "*(num-2),end="")
        print("*",end="")
        print("")
```
