```python
#Q-1
def greatest_num(n1, n2, n3):
    if n1>=n2 and n1>=n3:
        return f"{n1} is greatest"
    elif n2>=n1 and n2>=n3:
        return f"{n2} is greatest"
    else:
        return f"{n3} is greatest"
greatest_num(10, 20, 15)
```

```python
#Q-2
def celcius_to_fahrenheit(celcius):
    fahrenheit=(celcius * 9/5) + 32
    return f"{celcius} degree celcius is equal to {fahrenheit}"

celcius_to_fahrenheit(3)
```

```python
#Q-4
def sum_n(num):
    
    if num==1:
      return 1
    else:
      return num + sum_n(num-1)
print(sum_n(3))
```

```python
#Q-5
def pattern():
    num=int(input("Enter the number of rows: "))
    
    while num>0:
        print("*"*num)
        num-=1
pattern()    
```

```python
#Q-6
def inches_to_cm(inches):
    cm=inches*2.54
    return f"{inches} inches is equal to {cm} cm"

inches_to_cm(10)
```
